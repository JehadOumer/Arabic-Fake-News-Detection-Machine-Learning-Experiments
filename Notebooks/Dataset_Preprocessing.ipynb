```python
def Process(nlp, text):
    doc = nlp(text)
    text = replaceWordswithNER(doc, text)
    return text
```

```python
def printTokens(doc):
    for i, sentence in enumerate(doc.sentences):
        print('====== Sentence {i+1} tokens =======')
        for word in sentence.words:
            print(*[f'Word id: {word.id}\tToken id: {word.parent.id}\tSentence id: {i+1}\tToken: {word.parent.text}\ttext: {word.text}\tupos: {word.upos}\txpos: {word.xpos}\tlemma: {word.lemma}\tNER: {word.parent.ner}'],
                  sep='\n')
```

```python
skip_ner=["كورونا", "الكورونا", "الفايروس", "فايروس", "كوفيد" , "الكوفيد", "فيروس", "فيروس_كورونا", "كورونا فيروس", "فيروس كورونا", "لقاح_كورونا", "nCov", "COVID19 coronavirus"]
def replaceWordswithNER(doc, text):
  for sent in doc.sentences:
    for ent in sent.ents:
      if not any(ent.text in skip_item for skip_item in skip_ner):
        text = text.replace(ent.text, " " + ent.type+" ")
  return text
```

```python
nlp = stanza.Pipeline(lang="ar", processors='tokenize,mwt, ner')
full_dataset['tweet_text_NER'] = full_dataset['tweet_text'].map(lambda x: Process(nlp,x))
```
